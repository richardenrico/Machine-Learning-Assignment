```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
df = pd.read_csv('/content/heart_failure_clinical_records_dataset.csv')
df.head()
```

```python
df = df.loc[:, 'anaemia':'smoking']
df.drop(columns=['ejection_fraction'], inplace=True)
df
```

```python
df.loc[(df['creatinine_phosphokinase'] < 10) | (df['creatinine_phosphokinase'] > 120), 'creatinine_phosphokinase'] = 1
df.loc[(df['creatinine_phosphokinase'] >= 10) & (df['creatinine_phosphokinase'] <= 120), 'creatinine_phosphokinase'] = 0

df.loc[(df['platelets'] < 150000) | (df['platelets'] > 450000), 'platelets'] = 1
df.loc[(df['platelets'] >= 150000) & (df['platelets'] <= 450000), 'platelets'] = 0

df.loc[(df['serum_creatinine'] < 0.74) | (df['serum_creatinine'] > 1.35), 'serum_creatinine'] = 1
df.loc[(df['serum_creatinine'] >= 0.74) & (df['serum_creatinine'] <= 1.35), 'serum_creatinine'] = 0

df.loc[(df['serum_sodium'] < 135) | (df['serum_sodium'] > 145), 'serum_sodium'] = 1
df.loc[(df['serum_sodium'] >= 135) & (df['serum_sodium'] <= 145), 'serum_sodium'] = 0
```

```python
df.loc[df['creatinine_phosphokinase'] == 1]
```

```python
df1 = apriori(df, min_support=0.1, use_colnames=True)
df1
```

```python
rules = association_rules(df1, metric='lift', min_threshold=1)
rules
```
